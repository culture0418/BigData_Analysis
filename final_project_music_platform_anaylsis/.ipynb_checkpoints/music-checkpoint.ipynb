```python
# 匯入需要的模組
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['font.family'] = 'SimHei'
```

```python
# 匯入資料集
df = pd.read_excel('music.xlsx')
# 資料前處理 --- 篩掉不認真的回答、沒有使用過音樂平台
df = df[df['我不認真回答這份問卷?'] == '否']
df = df[df['你是否有使用過音樂串流平台(包含免費試用)?'] == '是']
df.columns
```

```python
# describe() 方法先來查看每個特徵的平均數、標準差、四分位數以及最大值與最小值
df.describe()
```

```python
# 資料前處理 -- 刪除填答秒數過短的資料(< 1/4 quater)
df = df[df["填答秒數"] > 64]
```

```python
# 資料前處理 -- 功能篩選
df[["人性化介面", "購買價格", "歌曲風格", "排行榜功能", "推薦歌曲", "有無MV", "有無歌曲demo", "同步"]].mean()
```

```python
# 資料前處理 -- 選取要分析的欄位數據
df = df[['年級', '每月可支配所得', '音樂串流平台', '最高價錢範圍', '語言', "曲風", "人性化介面", "購買價格", "歌曲風格", "同步"]]
df
```

```python
# 資料前處理 -- 資料轉換
def price(x):
    if x == "$0-$100/月":
        x = 0
    elif x == "$100-$150/月":
        x = 1
    elif x == "$150-$200/月":
        x = 2
    elif x == "$200-$250/月":
        x = 3
    elif x == "$250-$300/月":
        x = 4
    elif x == "$300以上/月":
        x = 5
    else:
        x = 10
    return int(x)

target_price = []

for i in df["最高價錢範圍"]:
    x = price(i)
    target_price.append(x)
```

```python
# 資料前處理 -- 資料轉換
def age(x):
    if x == "大一":
        x = 19
    elif x == "大二":
        x = 20
    elif x == "大三":
        x = 21
    elif x == "大四":
        x = 22
    elif x == "大五":
        x = 23
    elif x == "大六":
        x = 24
    else:
        x = 0
    return int(x)

target_age = []

for i in df["年級"]:
    x = age(i)
    target_age.append(x)
```

```python
df = df.drop(columns=["最高價錢範圍"])
df.insert(3, column = "最高價錢區間", value = target_price)

df = df.drop(columns=["年級"])
df.insert(0, column = "年齡(歲)", value = target_age)

df
```

```python
# 離散圖 : 曲風 & 音樂平台關係
plt.rcParams['axes.unicode_minus'] = False
df.plot(kind='scatter', title='音樂平台 & 曲風', figsize=(6,4), x='音樂串流平台', y='曲風', marker='+')

# 離散圖 : 價錢 & 音樂平台關係 
plt.rcParams['axes.unicode_minus'] = False
df.plot(kind='scatter', title='價錢 & 音樂平台', figsize=(6,4), x='音樂串流平台', y='最高價錢區間', marker='+')

# 離散圖 : 人性化界面 & 音樂平台關係
plt.rcParams['axes.unicode_minus'] = False
df.plot(kind='scatter', title = '人性化界面 & 音樂平台', figsize=(6,4), x = '音樂串流平台', y='人性化介面', marker='+')

# 離散圖 : 語言 & 音樂平台關係
plt.rcParams['axes.unicode_minus'] = False
df.plot(kind='scatter', title = '語言 & 音樂平台', figsize=(6,4), x = '音樂串流平台', y='語言', marker='+')
```

```python
# 圓餅圖 音樂平台佔有率
data_platform = df.loc[:, "音樂串流平台"]
data_kk = data_platform[data_platform == 'KKBOX'].count()
data_sp = data_platform[data_platform == 'SPOTIFY'].count()
data_youtube = data_platform[data_platform == 'YOUTUBE MUSIC'].count()
data_apple = data_platform[data_platform == 'APPLE MUSIC'].count()
data_street = data_platform[data_platform == 'STREETVOICE'].count()
data_others = data_platform[data_platform == '其他'].count()

data_circle_all_labels = [data_kk, data_sp, data_youtube, data_apple, data_street, data_others]
labels = ['KKBOX','SPOTIFY','YOUTUBE', 'APPLE MUSIC','STREETVOICE','其他']
plt.pie(data_circle_all_labels, labels=labels, autopct='%.1f%%')
plt.show()
```
