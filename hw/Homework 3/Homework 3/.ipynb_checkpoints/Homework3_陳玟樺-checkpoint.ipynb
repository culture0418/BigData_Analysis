```python
# 1-1
# Scikit-learn 網站中
# (https://scikit-learn.org/stable/datasets/toy_dataset.html)
# 有提供一些Toy datasets
# 其中，有個dataset為boston (波士頓房屋價格)。首先，請匯入boston 資料集。

from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import seaborn as sns
%matplotlib inline
font = FontProperties(fname='C:\Windows\Fonts\mingliu.ttc')

boston_data = datasets.load_boston() 
# 回傳格式為dictionary
# Key 值分別為 "data"、"target"、"target_names"、"DESCR"
boston_data
```

```python
#1-2
#秀出此資料集中共有幾個features，分別顯示feature_names。
boston_data['feature_names']
```

```python
# 1-2
len(boston_data['feature_names'])
```

```python
# 1-3
# 請用DataFrame方式將boston的data讀入X變數，並設定column的名稱為feature_names。
# 接著同樣用DataFrame方式將boston的target讀入Y變數，並設定column的名稱為「MEDV」。
# 最後利用concat函數將X與Y合併在同一個DataFrame中，
x = pd.DataFrame(boston_data['data'], columns=boston_data['feature_names'])
y = pd.DataFrame(boston_data['target'], columns=['MEDV'])
# 語法 : pd.contact( [x,y], axis = 1)
boston_data = pd.concat([x,y], axis=1) # 欄位合併
boston_data
```

```python
# 1-3 並秀出前五筆資料。
boston_data.head(5)
```

```python
# 1-4
# 請確認此資料集中是否有空缺值
# (亦即計算上一題的DataFrame中每個欄位的空缺值數目)。
boston_data.isnull().sum()    
```

```python
# 1-5
# 透過相關係數，可輔助迴歸模型之建立，
# 請秀出13個特徵變數(X變數)和Y變數(MEDV)的相關係數；
# 其中，相關係數請顯示至小數點第二位。

## 線性回歸，進行散步圖(相關係數)分析
## round(2) 小數點2位
boston_data1 = boston_data.corr()["MEDV"].round(2) 
boston_data1
```

```python
# 1-6
# 篩選自變數的基本方法可挑選相關係數較高的變數，
# 由上一題的結果中，請挑出X變數中與Y變數相關係數大於0.65的變數(不論正相關或負相關)，
# 將其設定為新的X變數，Y變數仍為MEDV。
# 接著使用sklearn中的函式把X、Y變數分為training及test datasets。
# 其中，training的資料占80 %、test 20%，並設定Random_state = 10，
# 並秀出 X_train的前五筆資料。

from sklearn.model_selection import train_test_split

x1 = boston_data[['LSTAT']]
x2 = boston_data[['RM']]
X = pd.concat([x1,x2], axis=1)
y = boston_data[['MEDV']]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=10) 

X_train.head(5)       
```

```python
# 1-7
# 承上題，請使用sklearn中的線性迴歸模式套件，
# 以training datasets的資料建立線性迴歸模型，
# 並分別秀出截距項、X變數之係數及R2

from sklearn import linear_model
#plt.style.use('ggplot')

regr = linear_model.LinearRegression()

regr.fit(X_train, y_train)

r_squared = regr.score(X_train, y_train)

print('截距項 = ', regr.intercept_)
print('X係數 = ', regr.coef_)
print('R平方 = ',r_squared)                                                                   
```

```python
# 1-8
# 衡量迴歸方程式好壞，常用的指標為均方根誤差(RMSE)，其公式如下：
# 請用test datasets計算均方根誤差(RMSE)，請顯示至小數點第四位，並秀出結果。
rmse = np.sqrt(((regr.predict(X_test) - y_test) ** 2). mean ( )).round(4)
print(len(rmse))
print(type(rmse))
print("均方根誤差 (root-mean-square error, RMSE) = " , rmse ) 
```

```python
# 1-9
# 衡量迴歸方程式好壞，另一個常用指標為平均絕對誤差(MAE)，其公式如下：
# 請用test datasets計算平均絕對誤差(MAE)，請顯示至小數點第四位，並秀出結果。
mae = (abs(regr.predict(X_test) - y_test).mean()).round(4)
print("平均絕對誤差(MAE) = ", mae)
```

```python
# 1-10
# 請使用第六題找出的X變數與Y變數(MEDV)，在分割為測試集與訓練集的情況下，
# 其中測試集佔i/100(i=1~99)，並設定Random_state = 10，
# 請找出測試集的均方根誤差RMSE值最小時的i，及其RMSE(請顯示至小數點第四位)
# (Note: 程式請使用迴圈，不要手動一次一次跑)
from sklearn import linear_model
from sklearn.model_selection import train_test_split


x1 = boston_data[['LSTAT']]
x2 = boston_data[['RM']]
X = pd.concat([x1,x2], axis=1)
y = boston_data[['MEDV']]
# X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=10) 
regr = linear_model.LinearRegression()

min_rmse = 5
rmse = 0
i_index = 0
for i in range(99): # i = 0 ~ 98
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = (i+1)/100, random_state=10) 
    regr.fit(X_train, y_train)
    
    rmse = np.sqrt(((regr.predict(X_test) - y_test) ** 2). mean ( )).round(4)
    rmse = float(rmse)
    # print(type(rmse))
    # print(min_rmse)
    if rmse < min_rmse:
        min_rmse = rmse
        i_index = i + 1
print("i = ", i_index)
print("min_rmse = ", min_rmse)
```

```python
# 1-11
# 同樣請使用第六題找出的X變數與Y變數(MEDV)，在分割為測試集與訓練集的情況下，
# 其中測試集佔i/100(i=1~99)，並設定Random_state = 10，
# 請找出測試集的R2最大時的i，及其R2 (請顯示至小數點第四位)
# (Note: 程式請使用迴圈，不要手動一次一次跑)
from sklearn import linear_model
from sklearn.model_selection import train_test_split


x1 = boston_data[['LSTAT']]
x2 = boston_data[['RM']]
X = pd.concat([x1,x2], axis=1)
y = boston_data[['MEDV']]

regr = linear_model.LinearRegression()

r_squared_max = 0
i_index = 0
for i in range(99): # i = 0 ~ 98
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = (i+1)/100, random_state=10) 
    regr.fit(X_train, y_train)
    
    r_squared = regr.score(X_train, y_train)
    
    if r_squared > r_squared_max:
        r_squared_max = r_squared
        i_index = i + 1
        

print("i = ", i_index)
print("r_squared_max = ", r_squared_max.round(4))
```

```python
# 1-12
# 若將13個特徵變數同時放入X變數，而Y變數為MEDV，
# 接著使用sklearn中的函式把X、Y變數分為training及test datasets；
# 其中，training的資料占80 %、test 20%，並設定Random_state = 10，
# 再使用sklearn中的線性迴歸模式套件，以training datasets的資料建立線性迴歸模型，
# 並分別秀出截距項、X變數之係數及R2。

from sklearn import linear_model # 回歸模式套件
from sklearn.model_selection import train_test_split

X = boston_data.iloc[:,0:13]
y = boston_data[['MEDV']]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=10) 

# 設定 linear regression 物件
regr_12 = linear_model.LinearRegression()

# 以 training datasets 的資料建立線性迴歸模型
regr_12.fit(X_train, y_train)

# 計算 R2
r_squared = regr_12.score(X_train, y_train)

print("截距 = ", regr_12.intercept_)
print("X變數之係數 = ", regr_12.coef_)
print("R平方 = ", r_squared)
```

```python
# 1-13
# 承上題，
# 假如13個特徵變數的值分別為0.00632、18、2.31、0、0.538、6.575、
# 65.2、4.09、1、296、15.3、396.9、4.98
# 請依據(12)題所建出的迴歸模型，預測房屋的中位數價格(MEDV)

x_variable = np.array([0.00632,18,2.31,0,0.538,6.575,65.2,4.09,1,296,15.3,396.9,4.98]).reshape((1, -1))
medv_predict = regr_12.predict(x_variable)

print("medv_predict = ", medv_predict)      
```
