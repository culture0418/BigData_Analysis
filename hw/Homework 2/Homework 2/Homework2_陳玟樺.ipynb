```python
#2-1
#請匯入health資料檔至DataFrame，並加以顯示。
import pandas as pd
import numpy as np

df = pd.read_excel("health.xlsx")
df
```

```python
#2-2
#在此DataFrame最後增加一個新的column
#計算「每日行走步數」除上「攝取卡路里」
#將此新的column名稱命名為「每單位卡路里之行走步數」，並加以顯示。
strides = df.iloc[:,1].values
cal = df.iloc[:,2].values
add_column = strides / cal
print(type(add_column))
df.insert(3, column = "每單位卡路里之行走步數", value = add_column, allow_duplicates=True)
df
```

```python
#2-3
#依據「每單位卡路里之行走步數」可判斷出每天的健康指數，
#若值小於等於2.5則顯示「低」，若值大於2.5、小於等於5則顯示「中」，若值大於5則顯示「高」，
#請在上題的DataFrame最後再增加一個新的column，顯示這些健康指數，並將此新的column名稱命名為「健康指數」。
bins = [0, 2.5, 5, 10]
index=['低','中','高']

df['健康指數']=pd.cut(df.iloc[:,3], bins, right=True, labels=index)
df      
```

```python
#2-4
#由於這些健康指數為數值型資料，使用One-hot Encoding方法將其擴展為三個columns 
#(亦即對應於該列的column為1，其餘columns為0)，將這三個新columns增加到上題DataFrame的最後，
#column名稱分別為「健康指數_高」，「健康指數_中」和「健康指數_低」。
onehot_encoding=pd.get_dummies(df['健康指數'], prefix='健康指數')
df = df.drop('健康指數',axis=1)
df_new = pd.concat([df,onehot_encoding],axis=1)
df_new         
```
