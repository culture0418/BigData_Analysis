```python
#2-6
print("請將Age欄位有空缺值的資料用「中位數」填補，並秀出填補後的DataFrame")
import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer
X = df2.iloc[:].values
#print(X.loc[:,"Age"].values)
#print(X.ndim)
#print(X[:, 5])

imp = SimpleImputer(missing_values = np.nan, strategy = 'median')
imp = imp.fit(X[:, 5:6])
#print(imp)

X[:, 5:6] = imp.transform(X[:, 5:6])
#print(X.loc[:,"Age"])
dataset2 = pd.DataFrame(X)

```

```python
#2-6
dataset2.columns=["PassengerId",  "Survived", "Pclass", "Name", "Sex",
                  "Age", "SibSp", "Parch", "Ticket", "Fare", "Embarked"]
dataset2
```

```python
#2-7
#rint("承上題，請繼續將Embarked欄位有空缺值之資料用「眾數」填補，並秀出填補Age與Embarked欄位後的DataFrame ")
import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer
X = dataset2.iloc[:].values

imp = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imp = imp.fit(X[:, 10:])
X[:, 10:] = imp.transform(X[:, 10:])

dataset2 = pd.DataFrame(X)
```

```python
#2-7
dataset2.columns=["PassengerId",  "Survived", "Pclass", "Name", "Sex",
                  "Age", "SibSp", "Parch", "Ticket", "Fare", "Embarked"]
dataset2
```

```python
#2-8
print("請以填補Age與Embarked欄位後的DataFrame，計算male(男生)和female(女生)的平均年齡，並加以顯示")
avg1 = dataset2[dataset2.loc[:, "Sex"] == "male"]
avg2 = dataset2[dataset2.loc[:, "Sex"] == "female"]

print("男生平均年齡 = {0} 歲".format(avg1.loc[:, "Age"].mean()))
print("女生平均年齡 = {0} 歲".format(avg2.loc[:, "Age"].mean()))
```

```python
#2-9
print("承上題，請列出男生和女生中的最年長及最年輕的年齡，並加以顯示\n")
avg1 = dataset2[dataset2.loc[:, "Sex"] == "male"] #男生
avg2 = dataset2[dataset2.loc[:, "Sex"] == "female"] #女生

print("男生最年輕 = {0} 歲\n".format(avg1.loc[:, "Age"].min()))
print("男生最年長 = {0} 歲\n".format(avg1.loc[:, "Age"].max()))
print("女生最年輕 = {0} 歲\n".format(avg2.loc[:, "Age"].min()))
print("女生最年長 = {0} 歲\n".format(avg2.loc[:, "Age"].max()))
```
