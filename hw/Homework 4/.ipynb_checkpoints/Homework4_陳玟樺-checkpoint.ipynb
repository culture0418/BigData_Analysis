```python
# 1-1 匯入 iris 資料(鳶尾花資料集)
from sklearn import datasets
import pandas as pd
import numpy as np
iris = datasets.load_iris()
iris
```

```python
# 1-2 
# 用 DataFrame 方式將 iris 的 data 讀入 X 變數
# 同時設定 column 的名稱為 feature_names，並秀出 X 的前五筆資料
X = pd.DataFrame(iris['data'], columns = iris['feature_names'])
X.head(5)
```

```python
# 1-3 
# 用 DataFrame 方式將 iris 的 target 讀入 Y 變數
# 同時設定 column 的名稱為「target」，並秀出 Y 的前五筆資料
Y = pd.DataFrame(iris['target'], columns = ['target'])
Y.head(5)
```

```python
# 1-4 請將 X、Y 合併在同一個 DataFrame 顯示 
data = pd.concat([X,Y], axis = 1)
data
```

```python
# 1-4 秀出此 DataFrame 的前五筆資料
data.head(5)
```

```python
# 1-5 
# 將此 DataFrame 的所有 features 全部保留，但在 target 部分請留下 0, 2 的部份，
# 請秀出更新後 DataFrame 的最後五筆資料
data_iris = data[data['target'].isin([0, 2])]
data_iris.tail(5)
```

```python
# 1-6 
# 為了使用感知器演算法，請新增一個 target_class 欄位
# 將 target 0 的部份對應成 1，target 2 的部份對應成 -1

target_class = (data_iris['target']).replace({0:1, 2:-1})
data_iris.insert(4, column = "target_class", value = target_class)
data_iris
```

```python
# 1-6 
# 並把原來的 target 那一欄刪除
# 請秀出更新後 DataFrame 的最後五筆資料

data_iris.pop("target")
data_iris.tail(5)
```

```python
# 1-7
def sign(z):
    if z > 0:
        return 1
    else:
        return -1
```

```python
# 1-7
# 請使用感知器演算法針對這兩種鳶尾花進行分類
# 其中起始 w0 設定為 0 向量(亦即數值全部設定為 0)
# 請秀出演算法 iteration 為幾次，最後所算出的 w為多少

w = np.array([0.,0.,0.,0.,0.])
error = 1
iterator = 0
while error != 0:
    error = 0
    for i in range(len(data_iris)):
        x,y = np.concatenate((np.array([1.]), np.array(data_iris.iloc[i])[:4])), np.array(data_iris.iloc[i])[4]
        if sign(np.dot(w,x)) != y:
            print("iterator: "+str(iterator))
            iterator += 1
            error += 1
            w += y*x            
            print("x: " + str(x))            
            print("w: " + str(w))  
```

```python
# 1-8
# 請再次使用感知器演算法針對這兩種鳶尾花進行分類，但將起始 w0 數值全部設定為 4
# 請再次秀出演算法 iteration 為幾次，最後所算出的 w 為多少。

w = np.array([4.,4.,4.,4.,4.])
error = 1
iterator = 0
while error != 0:
    error = 0
    for i in range(len(data_iris)):
        x,y = np.concatenate((np.array([1.]), np.array(data_iris.iloc[i])[:4])), np.array(data_iris.iloc[i])[4]
        if sign(np.dot(w,x)) != y:
            print("iterator: "+str(iterator))
            iterator += 1
            error += 1
            w += y*x            
            print("x: " + str(x))            
            print("w: " + str(w))
```

```python
# 1-9
# 假設現有個新的鳶尾花，四種 feature 分別為 [4.0, 3.1, 1.6, 1.3]
# 請利用第(8)題最後所計算出的 w，判斷這朵鳶尾花是 target 0 (Setosa)或 target 2 (Virginica)
print("w = {0}".format(w))
newflower = np.array([1, 4.0, 3.1, 1.6, 1.3]) 
print(sign(np.dot(w,newflower)))

def flower(a):
    if a == 1:
        print('target 0 (Setosa)')
    elif a == -1:
        print('target 2 (Virginica)')

print(flower(sign(np.dot(w,newflower))))  
```

```python
# 2-1 請匯入pokemon資料檔至DataFrame，並加以顯示
pk = pd.read_csv('pokemon.csv', encoding = 'big5')
pk
```

```python
# 2-2
# 寶可夢的Attack (攻擊力)、Defense(防禦力)兩欄位可能有遺漏值
# 請直接刪除這兩個欄位有遺漏值之寶可夢，並秀出刪除遺漏值後的資料表
pk = pk[pk['Attack'].notnull()]
pk = pk[pk['Defense'].notnull()]
pk
```

```python
# 2-3
# 承上題，Type(屬性)中包含許多屬性的寶可夢，
# 請僅留下屬性為 Fighting、Ghost 與 Normal 的寶可夢
# 並秀出更新後的資料表
pk = pk[pk['Type'].isin(['Fighting', 'Ghost', 'Normal'])]
pk
```

```python
# 2-4
# 請將 Attack (攻擊力)與 Defense(防禦力)這兩個欄位設定為 X 變數，
# 並進行變異數標準化(StandardScaler)後加以顯示

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = pk.loc[:,['Attack', 'Defense']]
sc.fit(X)
X = sc.transform(X)
# X = pd.DataFrame(X)
X
```

```python
# 2-5
# 因 Type (屬性)為類別欄位，請將屬性值進行編碼
# Fighting 對應成 0、Ghost 對應成 1、Normal 對應成 2
# 編碼完成後設定為 Y 變數，並加以顯示
Y = (pk['Type']).replace({'Fighting': 0, 'Ghost': 1, 'Normal': 2})
# Y = pd.DataFrame(Y)
Y
```

```python
# 2-6
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=0)
```

```python
# 2-6
# 請載入 SVM 中的 SVC，並將 kernel 設為線性(linear)(其他參數為預設值)，
# 使用上題(4)與(5)的 X(標準化後的)、Y 變數訓練此模型，
# 訓練完成後請與實際屬性值結果作比對，顯示出比對後的錯誤個數

from sklearn.svm import SVC
classifier = SVC(kernel='linear', probability=True)
classifier.fit(X_train, Y_train)
classifier.predict(X_test)

error = 0
# 有兩個參數, i控制實際值, v控制羅吉斯迴歸模型分類值, 逐一比對
for i, v in enumerate(classifier.predict(X_test)):
    if v!= Y_test.values[i]:
        error+=1
print(error)
```

```python
# 2-7
# 請進一步顯示出分類結果的混淆矩陣

from sklearn.metrics import confusion_matrix
Y_pred = classifier.predict(X_test)
confusion_matrix(Y_test, Y_pred)
```

```python
# 2-8
# 請計算分類結果的正確率(Accuracy)

from sklearn.metrics import accuracy_score
Y_pred=classifier.predict(X_test)
accuracy_score(Y_test, Y_pred)
```

```python
# 2-9
# 請計算分類結果的加權平均F1-score (請將average參數設定為weighted)
from sklearn.metrics import f1_score
y_pred=classifier.predict(X_test)
f1_score(y_test, y_pred, average='weighted')
```
